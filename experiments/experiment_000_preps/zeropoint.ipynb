```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
for survey in ["6dFGS", "SDSS", "LAMOST"]:
    params = np.load(f"../../artifacts/fp_fit/smin_setting_1/fp_fit_method_0/{survey.lower()}_chain.npy").T

    c_values = params[2] - params[0] * params[3] - params[1] * params[4]

    print(survey)
    print("Mean: ", np.mean(c_values))
    print("Std: ", np.std(c_values))
    print()

    plt.figure()
    plt.hist(c_values)
    plt.title(survey)
    plt.show()
```

# Data

```python
label = ["a", "b", "rmean", "smean", "imean", "sigma1", "sigma2", "sigma3", "c"]

for survey in ["6dFGS", "SDSS", "LAMOST"]:
    params = np.load(f"../../artifacts/fp_fit/smin_setting_1/fp_fit_method_0/{survey.lower()}_chain.npy").T

    c_values = params[2] - params[0] * params[3] - params[1] * params[4]

    params = np.append(params, [c_values], axis=0)

    print("=" * 25 + survey + "=" * 25)
    for i, param in enumerate(label):
        fp_param = params[i]
        mean_ = np.round(np.mean(fp_param), 3)
        std_ = np.round(np.std(fp_param), 3)
        
        print(f"{param}: ${mean_} \pm {std_}$")
    print()
```
